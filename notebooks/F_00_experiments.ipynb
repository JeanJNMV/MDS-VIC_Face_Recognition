```python
%matplotlib inline
from vic.dataloader import load_orl,make_fixed_test_indices,split_with_fixed_test
from vic.models import Eigenfaces
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
data = load_orl("../data/ORL")
test_idx, pool_idx = make_fixed_test_indices(data, n_test=3, seed=0)
```

```python
Xtr, ytr, Xte, yte = split_with_fixed_test(data, test_idx, pool_idx, 2, seed=0)
print(f"Train size: {Xtr.shape[0]}, Test size: {Xte.shape[0]}")
```

```python
eigen_faces = Eigenfaces(K=10)
```

```python
eigen_faces.fit(Xtr, ytr)
```

```python
y_pred = eigen_faces.predict(Xte)
```

```python
accuracy_score(yte, y_pred)
```
