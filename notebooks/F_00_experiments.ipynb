```python
%matplotlib inline
from vic import dataloader, models
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
imgs, ids = dataloader.load_ORL("../data/ORL")
```

```python
N, H, W = imgs.shape

imgs_r = imgs.reshape(-1, 5, H, W)
labels_r = ids.reshape(-1, 5)

train_imgs   = imgs_r[:, :3].reshape(-1, H, W)
train_labels = labels_r[:, :3].reshape(-1)

test_imgs    = imgs_r[:, 3:].reshape(-1, H, W)
test_labels  = labels_r[:, 3:].reshape(-1)

train_imgs.shape, test_imgs.shape
```

```python
eigen_faces = models.EigenFaces(200)
```

```python
eigen_faces.fit(train_imgs, train_labels)
```

```python
y_pred = eigen_faces.predict(test_imgs)
```

```python
accuracy_score(test_labels, y_pred)
```
